### 확률분포 가정하기
#### -> 먼저, 히스토그램을 통해 모양을 관찰한다.

#### 1. 데이터가 2개의 값(0 또는 1)만 가지는 경우 : 베르누이 분포
#### 2. 데이터가 n개의 이산적인 값을 가지는 경우: 카테고리 분포
#### 3. 데이터가 [0,1] 사이에서 값을 가지는 경우: 베타 분포
#### 4. 데이터가 0 이상의 값을 가지는 경우: 감마분포, 로그정규분포 등
#### 5. 데이터가 R(실수) 전체에서 값을 가지는 경우: 정규분포, 라플라스분포

# 베르누이 분포 (0 또는 1)

```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

bern_data = np.random.binomial(1, 0.5, 1000)
plt.hist(bern_data, bins=2, edgecolor='k')
plt.title('베르누이 분포')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 카테고리 분포 (이산적인 값 n개)


```python
cat_data = np.random.choice([0, 1, 2, 3, 4], size=1000, p=[0.1, 0.2, 0.3, 0.2, 0.2])
plt.hist(cat_data, bins=5, edgecolor='k')
plt.title('카테고리 분포')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 베타 분포 (0~1 사이의 값)

```python
beta_data = np.random.beta(a=2, b=5, size=1000)
plt.hist(beta_data, bins=20, edgecolor='k')
plt.title('베타 분포')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 감마 분포 (0 이상의 값)

```python
gamma_data = np.random.gamma(shape=2, scale=1, size=1000)
plt.hist(gamma_data, bins=20, edgecolor='k')
plt.title('감마 분포')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 로그 정규분포 (0 이상의 값)

```python
lognorm_data = np.random.lognormal(mean=0, sigma=1, size=1000)
plt.hist(lognorm_data, bins=20, edgecolor='k')
plt.title('로그정규분포')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 라플라스분포 (실수 전체)

```python
laplace_data = np.random.laplace(loc=0, scale=1, size=1000)
plt.hist(laplace_data, bins=20, edgecolor='k')
plt.title('라플라스 분포')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
